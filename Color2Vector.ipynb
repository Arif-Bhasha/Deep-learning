```python
video1=cv2.VideoCapture(0)
count=0
while True:
    count=count+1
    su,img1=video1.read()      #obj (video1) has attribute read- it read frame by frame
    cv2.imshow('camera',img1)  #showing each frame
    cv2.imwrite(r'C:\Users\kalali aravindgoud\Desktop\Dataset\1-cv2\{}.jpg'.format(count),img1) #saving each frame
    if cv2.waitKey(1)==ord('a'): #ascii value
        break
cv2.destroyAllWindows()
```
